```python
import psycopg2
from prettytable import PrettyTable as pt
```

```python
class karthi:
    def connection(self):
        con=psycopg2.connect(host='localhost',database='empdpt',user='postgres',password='karthi')
        return con
    
    def inner_join(self):
        if (self.connection()):
            con=self.connection()
            cur=con.cursor()
            sql=''' select emp.id,emp.emp_id,emp.name,dept.dept from emp inner join dept on emp.emp_id=dept.emp_id;'''
            cur.execute(sql)
            data=cur.fetchall()
            t = Pt(['id', 'emp_id','name','dept'])
            for data in data:
                t.add_row(data)
            print(t)
            con.close()

    def left_join(self):
        if (self.connection()):
            con=self.connection()
            cur=con.cursor()
            sql=''' select emp.id,emp.emp_id,emp.name,dept.dept from emp left join dept on emp.emp_id=dept.emp_id;'''
            cur.execute(sql)
            data=cur.fetchall()
            t = Pt(['id', 'emp_id','name','dept'])
            for data in data:
                t.add_row(data)
            print(t)
            con.close()

    def right_join(self):
        if (self.connection()):
            con=self.connection()
            cur=con.cursor()
            sql=''' select emp.id,emp.emp_id,emp.name,dept.dept from emp right join dept on emp.emp_id=dept.emp_id;'''
            cur.execute(sql)
            data=cur.fetchall()
            t = Pt(['id', 'emp_id','name','dept'])
            for data in data:
                t.add_row(data)
            print(t)
            con.close()

    def cross_join(self):
        if (self.connection()):
            con=self.connection()
            cur=con.cursor()
            sql=''' select * from emp cross join dept;'''
            cur.execute(sql)
            data=cur.fetchall()
            t = Pt(['id','name','emp_id','dept.id','dept','dept.emp_id'])
            for data in data:
                t.add_row(data)
            print(t)
            con.close()
    
    def full_join(self):
        if (self.connection()):
            con=self.connection()
            cur=con.cursor()
            sql=''' select emp.id,emp.emp_id,emp.name,dept.dept from emp full join dept on emp.emp_id=dept.emp_id;'''
            cur.execute(sql)
            data=cur.fetchall()
            t = Pt(['id', 'emp_id','name','dept'])
            for data in data:
                t.add_row(data)
            print(t)
            con.close()


k=karthi()
print("""Enter 1 to inner join
        Enter 2 to left join
        Enter 3 to right join
        Enter 4 to cross join
        Enter 5 to full join""")
n = int(input("Enter the choice:"))
while True:
    if(n==1):
        k.inner_join()
    elif(n==2):
        k.left_join()
    elif(n==3):
        k.right_join()
    elif(n==4):
        k.cross_join()
    elif(n==5):
        k.full_join()
    else:
        break

    
```
